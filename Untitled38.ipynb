```python
import pandas as pd

train = pd.read_csv('application_train.csv')
test = pd.read_csv('application_test.csv')

print(train.shape)
print(train['TARGET'].value_counts(normalize=True))
```

```python
features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH', 'DAYS_EMPLOYED']
X = train[features]
y = train['TARGET']
X_test = test[features]

# Optional: handle missing values, scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

val_preds = model.predict_proba(X_val)[:, 1]
auc_score = roc_auc_score(y_val, val_preds)
print(f'Validation AUC: {auc_score:.4f}')
```

```python
test_preds = model.predict_proba(X_test_scaled)[:, 1]
submission = pd.DataFrame({
    'SK_ID_CURR': test['SK_ID_CURR'],
    'TARGET': test_preds
})
submission.to_csv('baseline_submission.csv', index=False)
```

```python
import numpy as np # Import numpy

for col in ['AMT_INCOME_TOTAL', 'AMT_CREDIT']:
    train[col + '_LOG'] = np.log1p(train[col])
    test[col + '_LOG'] = np.log1p(test[col])

features = ['AMT_INCOME_TOTAL_LOG', 'AMT_CREDIT_LOG', 'DAYS_BIRTH', 'DAYS_EMPLOYED']
X = train[features]
X_test = test[features]
```

```python
X_scaled = StandardScaler().fit_transform(X)
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model.fit(X_train, y_train)
auc4 = roc_auc_score(y_val, model.predict_proba(X_val)[:, 1])
```

```python
train['CREDIT_INCOME_RATIO'] = train['AMT_CREDIT'] / train['AMT_INCOME_TOTAL']
test['CREDIT_INCOME_RATIO'] = test['AMT_CREDIT'] / test['AMT_INCOME_TOTAL']
train['AMT_INCOME_TOTAL_LOG'] = np.log1p(train['AMT_INCOME_TOTAL'])
test['AMT_INCOME_TOTAL_LOG'] = np.log1p(test['AMT_INCOME_TOTAL'])

features = ['AMT_INCOME_TOTAL_LOG', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'CREDIT_INCOME_RATIO']
X = train[features]
X_test = test[features]
```

```python
X_scaled = StandardScaler().fit_transform(X)
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model.fit(X_train, y_train)
auc5 = roc_auc_score(y_val, model.predict_proba(X_val)[:, 1])
```

```python
final_preds = model.predict_proba(StandardScaler().fit_transform(X_test))[:, 1]
submission = pd.DataFrame({
    'SK_ID_CURR': test['SK_ID_CURR'],
    'TARGET': final_preds
})
submission.to_csv('final_submission.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style('whitegrid')

# Define the features to visualize
features_to_visualize = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH', 'DAYS_EMPLOYED']

# Create subplots for each feature
fig, axes = plt.subplots(len(features_to_visualize), 1, figsize=(10, 5 * len(features_to_visualize)))

# Plot histograms for each feature
for i, feature in enumerate(features_to_visualize):
    sns.histplot(data=train, x=feature, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Visualize the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(data=train, x='TARGET')
plt.title('Distribution of Target Variable')
plt.xlabel('TARGET (0: No Difficulty, 1: Difficulty)')
plt.ylabel('Count')
plt.show()
```
