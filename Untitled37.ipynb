```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

class CNN_Encoder(tf.keras.Model):
    def __init__(self, embed_dim):
        super(CNN_Encoder, self).__init__()
        self.resnet = ResNet50(include_top=False, weights='imagenet')
        self.global_pool = GlobalAveragePooling2D()
        self.fc = Dense(embed_dim, activation='relu')

    def call(self, x):
        x = self.resnet(x)
        x = self.global_pool(x)
        x = self.fc(x)
        return x  # shape: (batch_size, embed_dim)
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense, Concatenate, RepeatVector

class RNN_Decoder(tf.keras.Model):
    def __init__(self, vocab_size, embed_dim, units, max_len):
        super(RNN_Decoder, self).__init__()
        self.embedding = Embedding(vocab_size, embed_dim, mask_zero=True)
        self.repeat = RepeatVector(max_len)
        self.concat = Concatenate(axis=-1)
        self.lstm = LSTM(units, return_sequences=True)
        self.fc = Dense(vocab_size, activation='softmax')

    def call(self, image_features, captions):
        caption_embed = self.embedding(captions)
        image_embed = self.repeat(image_features)
        x = self.concat([image_embed, caption_embed])
        x = self.lstm(x)
        x = self.fc(x)
        return x  # shape: (batch_size, max_len, vocab_size)
```

```python
class ImageCaptioningModel(tf.keras.Model):
    def __init__(self, vocab_size, embed_dim, units, max_len):
        super(ImageCaptioningModel, self).__init__()
        self.encoder = CNN_Encoder(embed_dim)
        self.decoder = RNN_Decoder(vocab_size, embed_dim, units, max_len)

    def call(self, image, caption):
        image_features = self.encoder(image)
        output = self.decoder(image_features, caption)
        return output
```

```python
def generate_caption(model, image, tokenizer, max_len):
    caption_seq = [tokenizer.word_index['<start>']]
    for _ in range(max_len):
        padded = tf.keras.preprocessing.sequence.pad_sequences([caption_seq], maxlen=max_len)
        preds = model(image, padded)
        next_word = tf.argmax(preds[0, len(caption_seq)-1]).numpy()
        caption_seq.append(next_word)
        if next_word == tokenizer.word_index['<end>']:
            break
    return tokenizer.sequences_to_texts([caption_seq])[0]
```

```python
import tensorflow as tf

class CaptionTrainer(tf.keras.Model):
    def __init__(self, caption_model):
        super().__init__()
        self.caption_model = caption_model
        self.loss_tracker = tf.keras.metrics.Mean(name="loss")
        self.acc_tracker = tf.keras.metrics.SparseCategoricalAccuracy(name="accuracy")

    def compile(self, optimizer, loss_fn):
        super().compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn

    def train_step(self, data):
        images, captions_in, captions_out = data

        with tf.GradientTape() as tape:
            predictions = self.caption_model(images, captions_in)
            loss = self.loss_fn(captions_out, predictions)

        grads = tape.gradient(loss, self.caption_model.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.caption_model.trainable_variables))

        self.loss_tracker.update_state(loss)
        self.acc_tracker.update_state(captions_out, predictions)
        return {"loss": self.loss_tracker.result(), "accuracy": self.acc_tracker.result()}
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False, reduction='none')

def masked_loss(y_true, y_pred):
    loss = loss_fn(y_true, y_pred)
    mask = tf.cast(tf.not_equal(y_true, 0), dtype=loss.dtype)
    loss *= mask
    return tf.reduce_mean(loss)
```
