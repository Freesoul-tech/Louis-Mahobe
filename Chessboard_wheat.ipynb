```python
#Wheat and chess board problem by Louis,,,,
import numpy as np

chessboard = np.zeros((2,2), dtype = int)
chessboard [0,0] =1
chessboard [0,1] =2
chessboard [1,0] =4
chessboard [1,1] =8
print(chessboard)

def create_chessboard(n,m):
  chessboard = np.zeros((n,m), dtype = np.uint)
  for i in range(n):
    for j in range(m):
      chessboard[i,j] = 2**(i*m +j)
  return chessboard
chessboard_8x8 = create_chessboard(8, 8)
print(chessboard_8x8)

import matplotlib.pyplot as plt

#  number of wheat
total_wheat = np.sum(chessboard_8x8)
print("Total number of wheat:", total_wheat)

average_per_column = np.mean(chessboard_8x8, axis=0)

plt.xlabel("Column")
plt.ylabel("Average number of wheat")
plt.title("Average number of wheat in each column")
plt.bar(np.arange(1, 9), average_per_column)
plt.show()

plt.xlabel("Column")
plt.ylabel("Row")
plt.title("Heatmap of Wheat on Chessboard")
plt.pcolor(chessboard_8x8)
plt.colorbar()
plt.show()

first_half = chessboard_8x8[:4, :].sum()
second_half = chessboard_8x8[4:, :].sum()

ratio = second_half / first_half
print("Ratio of wheat in the second half to the first half:", ratio)

def create_chessboard_append(n, m):
    n_squares = n * m
    chessboard = np.array([1], dtype=np.uint64)
    for _ in range(n_squares - 1):
        chessboard = np.append(chessboard, 2 * chessboard[-1])
    return chessboard.reshape((n, m))

chessboard_append = create_chessboard_append(8, 8)
print(chessboard_append)

def create_chessboard_broadcast(n, m):
    n_squares = n * m
    indices_of_squares = np.arange(n_squares, dtype=np.uint64)
    chessboard = 2**indices_of_squares
    return chessboard.reshape((n, m))

chessboard_broadcast = create_chessboard_broadcast(8, 8)
print(chessboard_broadcast)

import time

#  measure time
def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

#  execution times
time_chessboard = measure_time(create_chessboard, 8, 8)
time_chessboard_append = measure_time(create_chessboard_append, 8, 8)
time_chessboard_broadcast = measure_time(create_chessboard_broadcast, 8, 8)

print("Execution time of create_chessboard:", time_chessboard)
print("Execution time of create_chessboard_append:", time_chessboard_append)
print("Execution time of create_chessboard_broadcast:", time_chessboard_broadcast)


```
