```python
import pandas as pd

# Load data
df = pd.read_csv('/application_train.csv')
desc = pd.read_csv('/HomeCredit_columns_description.csv', encoding='ISO-8859-1')

# Basic inspection
print(df.head())        # Preview first few rows
print(df.info())        # Data types & missing values
print(df.describe())    # Statistical summary
```

```python
# Check missing value ratio
missing_ratio = df.isnull().mean().sort_values(ascending=False)

# Impute with median (excluding outliers)
for col in df.columns:
    if df[col].isnull().sum() > 0 and df[col].dtype != 'object':
        median = df[col].median()
        df[col] = df[col].fillna(median)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df.select_dtypes(include=['number']).corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart of TARGET distribution
target_counts = df['TARGET'].value_counts()
labels = ['Repaid', 'Defaulted']
colors = ['#66b3ff','#ff9999']

plt.figure(figsize=(6,6))
plt.pie(target_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Loan Repayment vs Default')
plt.axis('equal')
plt.show()
```

```python
features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH']

plt.figure(figsize=(15,5))
for i, feature in enumerate(features):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[feature], bins=50, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```
