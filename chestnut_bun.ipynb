```python
import numpy as np


chestnut_bun_volume = (4/3) * np.pi * (3**3)

# Solar system volume
solar_system_volume = (4/3) * np.pi * (4.545e+14)**3
import matplotlib.pyplot as plt

def time_to_fill(initial_volume, target_volume, doubling_time):

    time = 0
    current_volume = initial_volume
    volumes = [current_volume]
    times = [time]

    while current_volume < target_volume:
        time += doubling_time
        current_volume *= 2
        volumes.append(current_volume)
        times.append(time)

    #  volume growth
    plt.plot(times, volumes)
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume (cm^3)")
    plt.title("Volume Growth ")
    plt.grid(True)
    plt.show()

    return time

time = time_to_fill(chestnut_bun_volume, solar_system_volume, 5)

print(f"It will take {time} minutes for the chestnut bun to fill the solar system.")
```
