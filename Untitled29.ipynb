```python
import numpy as np

def initialize_centers(X, K):
    indices = np.random.choice(len(X), K, replace=False)
    return X[indices]
```

```python
def calculate_sse(X, labels, centers):
    sse = 0.0
    for k in range(len(centers)):
        cluster_points = X[labels == k]
        sse += np.sum((cluster_points - centers[k])**2)
    return sse
```

```python
def assign_clusters(X, centers):
    distances = np.linalg.norm(X[:, np.newaxis] - centers, axis=2)
    return np.argmin(distances, axis=1)
```

```python
def update_centers(X, labels, K):
    new_centers = np.array([X[labels == k].mean(axis=0) for k in range(K)])
    return new_centers
```

```python
def kmeans_single_run(X, K, max_iter=100, tol=1e-4):
    centers = initialize_centers(X, K)
    for _ in range(max_iter):
        labels = assign_clusters(X, centers)
        new_centers = update_centers(X, labels, K)
        if np.linalg.norm(new_centers - centers) <= tol:
            break
        centers = new_centers
    sse = calculate_sse(X, labels, centers)
    return centers, labels, sse
```

```python
def kmeans(X, K, n_init=10, max_iter=100, tol=1e-4):
    best_sse = np.inf
    best_centers, best_labels = None, None
    for _ in range(n_init):
        centers, labels, sse = kmeans_single_run(X, K, max_iter, tol)
        if sse < best_sse:
            best_sse = sse
            best_centers, best_labels = centers, labels
    return best_centers, best_labels, best_sse
```

```python
def predict(X_new, centers):
    distances = np.linalg.norm(X_new[:, np.newaxis] - centers, axis=2)
    return np.argmin(distances, axis=1)
```

```python
import matplotlib.pyplot as plt

def elbow_method(X, max_k=10):
    sses = []
    for k in range(1, max_k + 1):
        _, _, sse = kmeans(X, k)
        sses.append(sse)
    plt.plot(range(1, max_k + 1), sses, marker='o')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('SSE')
    plt.title('Elbow Method')
    plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

def silhouette_analysis(X, labels):
    silhouette_vals = silhouette_samples(X, labels)
    silhouette_avg = silhouette_score(X, labels)
    return silhouette_vals, silhouette_avg
```
