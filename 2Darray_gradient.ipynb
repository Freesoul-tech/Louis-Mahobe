```python
# Two-dimension arrays and gradient

import numpy as np

x = np.arange(-50, 50.1, 0.1)  # ndarray representing x
y = 0.5 * x + 1  # ndarray representing y
combined_array = np.column_stack((x, y))  # combine x and y into a 2D array
print(combined_array)
gradient= np.diff(y)/np.diff(x)
print(gradient)

import matplotlib.pyplot as plt

# Function graph
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='y = 0.5x + 1')
plt.title('First-Order Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

# Gradient graph
x_gradient = x[:-1]  # Adjusted x for gradient
plt.figure(figsize=(10, 5))
plt.plot(x_gradient, gradient, label='Gradient', color='orange')
plt.title('Gradient of First-Order Function')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.legend()
plt.grid()
plt.show()

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)  # Create x array
    y = function(x)  # Compute y array
    gradient = np.diff(y) / np.diff(x)  # Gradient calculation
    combined_array = np.column_stack((x, y))  # Combine x and y
    return combined_array, gradient

def function1(array_x):
    return 0.5 * array_x + 1

#  usage
array_xy, gradient = compute_gradient(function1)
def function2(array_x):
    return array_x ** 2

def function3(array_x):
    return 2 * array_x ** 2 + 2 * array_x

def function4(array_x):
    return np.sin(array_x / 2)

# Gradients
_, gradient1 = compute_gradient(function2)
_, gradient2 = compute_gradient(function3)
_, gradient3 = compute_gradient(function4, x_range=(0, 50.1, 0.1))

y_min = np.min(y)
x_min_index = np.argmin(y)
gradient_before = gradient[x_min_index - 1] if x_min_index > 0 else None
gradient_after = gradient[x_min_index] if x_min_index < len(gradient) else None

print(f"Minimum y: {y_min}")
print(f"Slope before: {gradient_before}")
print(f"Slope after: {gradient_after}")

```
